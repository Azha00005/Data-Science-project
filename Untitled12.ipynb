```python
import pandas as pd
import numpy as np

# Створюємо набір даних згідно з Варіантом 1
data = {
    'Кредитна історія': ['погана', 'невідома', 'невідома', 'невідома', 'невідома', 'невідома', 'погана', 'погана', 'гарна', 'гарна', 'гарна', 'гарна', 'гарна', 'погана'],
    'Борг': ['високий', 'високий', 'низький', 'низький', 'низький', 'високий', 'низький', 'низький', 'низький', 'високий', 'високий', 'високий', 'високий', 'високий'],
    'Порука': ['немає', 'немає', 'немає', 'немає', 'немає', 'адекватне', 'немає', 'адекватне', 'немає', 'адекватне', 'немає', 'немає', 'немає', 'немає'],
    'Дохід': ['0-15', '0-15', '0-15', '0-15', '>35', '>35', '0-15', '>35', '>35', '>35', '0-15', '0-15', '>35', '0-15'],
    'Ризик': ['високий', 'високий', 'середній', 'високий', 'низький', 'низький', 'високий', 'середній', 'низький', 'низький', 'високий', 'середній', 'низький', 'високий']
}

df = pd.DataFrame(data)
```

Функції для розрахунку Ентропії та Gain

```python
def calculate_entropy(y):
    # Формула: - sum(pi * log2(pi))
    probs = y.value_counts(normalize=True)
    return -np.sum(probs * np.log2(probs))

def calculate_gain(df, attribute, target_name):
    # Початкова ентропія всієї таблиці [cite: 77]
    total_entropy = calculate_entropy(df[target_name])

    # Середня ентропія після поділу за атрибутом [cite: 78]
    values = df[attribute].unique()
    weighted_entropy = 0
    for val in values:
        subset = df[df[attribute] == val]
        prob = len(subset) / len(df)
        weighted_entropy += prob * calculate_entropy(subset[target_name])

    # Формула (3.2): Виграш інформації
    return total_entropy - weighted_entropy
```

Знаходження кореня дерева

```python
target = 'Ризик'
attributes = ['Кредитна історія', 'Борг', 'Порука', 'Дохід']

gains = {attr: calculate_gain(df, attr, target) for attr in attributes}
for attr, gain in gains.items():
    print(f"Gain({attr}) = {gain:.4f}")

root_node = max(gains, key=gains.get)
print(f"\nНайкращий атрибут для кореня: {root_node}")
```

Візуалізація дерева рішень

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Кодуємо текстові дані в числа
le = LabelEncoder()
df_numeric = df.apply(le.fit_transform)

X = df_numeric.drop('Ризик', axis=1)
y = df_numeric['Ризик']

# Створюємо модель дерева
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

# Малюємо дерево
plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=X.columns, class_names=le.classes_, filled=True)
plt.show()
```
